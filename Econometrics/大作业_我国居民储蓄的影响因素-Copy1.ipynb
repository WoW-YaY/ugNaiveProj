## 一、引言

### （一）问题提出

改革开放后，中国居民储蓄迎来了“井喷式”的增长。根据央行统计数据显示，2020年一季度末中国居民存款一共增加了6.47万亿元(去年同期为6.07万亿)，截止2020年一季度末，中国居民存款余额达87.8万亿元，人均存款6.27万元，户均存款16.93万元。这意味着，中国仍旧是全球居民储蓄率最高的国家之一。同时据中金公司首席经济学家彭文生发布的《中金2020下半年宏观经济展望》报告，新冠疫情期间中国居民储蓄率大幅上升。但是，随着中国经济呈现高质量发展态势，储蓄率下降无法避免，储蓄率下降更是一个长期趋势。根据国际货币基金组织的研究，由于人口结构变化和政策导向，预计中国居民储蓄占GDP比重到2030年将下降近10个百分点。鉴于居民储蓄对于经济发展的重要作用，分析影响居民储蓄的因素显得尤为关键。目前该领域的研究已经有充分的理论基础，但是一直缺乏系统性的实证分析。本课题即试图基于前人定性的理论，进一步给出数量上定量的分析。

### （二）研究意义

居民储蓄对国家经济有一定影响，尤其像我们国家，经济的发展模式是由政府主导的大规模投资和大规模出口，而其中投资的资金大部分来源于银行，银行的资金来源于居民的储蓄。另一方面，过高的储蓄会导致内需不足、消费下降，企业的生产积极性降低，不利于企业进行再生产，进而对经济发展产生不良影响。可见，研究影响居民储蓄的因素，并提出相应的政策性建议使得居民储蓄长期运行在平稳区间，对于国家经济的长远发展来说，正当其时、十分必要。

### （三）文献综述

在对于我国居民储蓄影响因素的研究方面，有一些学者做出了实证研究的贡献。李美华[1]选取了对居民储蓄的变化影响最大的两个因素来进行实证分析，分别是国民总收入和居民消费价格指数。他的模型证明了，必须保证居民收入和消费两者的共同发展，才能保证居民储蓄存款的稳步增加。他的突出贡献在于，进一步分析了国民总收入和居民储蓄率的相互作用机制，发现居民储蓄存款余额过高会反过来对银行等金融机构的发展带来负面影响，阻碍经济的发展和国民总收入的快速增长，进而提出要刺激消费的政策建议。许奥[2]选取了人均可支配收入、利率水平和通货膨胀水平三个因素（还有一个居民对未来收入的预期程度，可能由于难以量化，作者并未放入模型中），通过建立双对数模型来反映这三个方面对于居民储蓄的影响程度，并提出了相应的建议以降低储蓄率，刺激消费拉动经济增长。余新宏等[3]则额外选择了S&P股票指数，并发现其与居民的储蓄率成正相关。这与日常逻辑是相悖的，因为当居民的储蓄增加时，投资到股票市场的资金会相对会减少，因而它们应该存在负相关的关系。他们认为出现该结论的原因可能是我国的资本市场建立还不是很完善，存在着较大的股票投机行为。

## 二、理论分析与研究思路

### （一）理论分析

首先我们从理论上探讨居民储蓄率可能的影响因素。

根据凯恩斯的理论，收入是居民储蓄的最重要决定因素，储蓄是收入的增函数，并且边际储蓄倾向（储蓄函数的斜率）是上升的。因为居民储蓄是居民把可支配收入中暂时不用于消费的部分存入银行，因此可支配收入是储蓄的源泉。但是正因为储蓄余额随着可支配收入的提高几乎是必然增加的，所以这样的研究无助于居民储蓄意愿的探讨。我们应该从更多的角度探究，在什么样的经济环境下居民愿意把占收入多大比例的钱存起来。因此比较合适的方式是把居民储蓄率作为被解释变量。同时根据哈罗德-多马模型，$经济增长率 = {储蓄率\over 资本} - 产出比率$，可以发现其实储蓄率才是拉动经济增长的关键。

简单来说，居民把钱用于消费还是储蓄，大体上受两股力量的影响，一是满足日常开支要花多少钱，二是把钱存起来有多少吸引力。这两个因素分别可以用居民消费价格指数和银行存款利率来衡量。直观上来看，在可支配收入不变的情况下，居民消费价格指数越高，消费开支占可支配收入的比重越高，储蓄率就越低，银行存款利率越高，存款占可支配收入的比重越高，储蓄率就越高。同时，居民消费价格指数也是衡量通货膨胀情况的重要参考。而更能影响居民直观消费感受的应该是通货膨胀的速度，即通货膨胀率。

另一方面，受疫情期间的观察所启发，社会失业率的提升可能使居民对收入水平产生悲观预期，可能加大储蓄的比重以应对未知的风险。因此，失业率越高，储蓄率也应该越高。

因此，通货膨胀率、银行存款利率和失业率可能是影响居民储蓄率的主要因素，其中通货膨胀率与居民储蓄率成负相关，银行存款利率与居民储蓄率成正相关，失业率与居民储蓄率成正相关。从数值上来说，居民储蓄率、通货膨胀率、存款利率和失业率都是百分比，从经济意义上来说，目前没有理论表明通货膨胀率、银行存款利率和失业率对于储蓄率的影响有边际递减的效应，所以我们猜想一般的线性模型可能已经适用。

### （二）研究思路

确定研究对象是一切研究的开始。本课题首先从实践意义的角度给出了被解释变量——居民储蓄率，再从一般经济学的定性理论出发，给出影响储蓄率的主要因素，分别是通货膨胀率、银行存款利率和失业率，并探讨这些解释变量各自与储蓄率的定性关系。然后设定回归模型的形式，再根据搜集到的数据对模型进行估计，之后按照一般方法对模型的估计结果进行统计学和经济学检验，依照检验的结果对模型进行修改（如有必要）。最后得出被解释变量与解释变量之间的定量关系，并以此来给出政策性建议。

## 三、计量经济模型与估计方法

### （一）模型设定

设定计量经济模型，首先就是确定模型中的变量。正如上面的理论分析中指出的，我们根据研究的目的，即居民储蓄的主观意愿的探讨，确定以居民储蓄率作为被解释变量，根据经济理论为指导，确定以通货膨胀率、银行存款利率和失业率作为解释变量。

回归模型的设定除了选择模型中的变量外，还要使所设定的变量间函数形式能够体现变量间的基本关系，即根据对变量间相互关系的已有认识，把$Y$的条件期望设定为解释变量$X$的某种函数。总体条件期望$E(Y|X_i) = f(X_i)$所设定的函数形式要体现变量间的基本经济关系。而过于复杂的模型可能难以从经济意义的角度作出解释，从上面的理论分析中也可以看出，没有证据表明有设定更复杂模型的必要，因此考虑暂定变量间的关系为线性关系，即假设解释变量每变动1个单位，被解释变量会相应变动一定单位。

我们以居民储蓄率为被解释变量$Y$，通货膨胀率为解释变量$X_2$、银行存款利率为解释变量$X_3$，失业率为解释变量$X_4$。建立多元线性回归模型

$$Y_i = \beta_1 + \beta_2X_{2i} + \beta_3X_{3i} + \beta_4X_{4i} + u_i$$

其中$\beta_1, \beta_2, \beta_3, \beta_4$是常数，而通货膨胀率与居民储蓄率成负相关，银行存款利率与居民储蓄率成正相关，失业率与居民储蓄率成正相关，因此应有$\beta_2<0, \beta_3>0, \beta_4>0$。

### （二）估计方法

在确定了模型的基本形式后，就需要用所搜集的数据对模型中的参数进行估计。这里我们选用的是OLS（普通最小二乘法），因为这种方法不仅简洁易用，而且在很多情况下都是既简便又适用的估计方法。在多元线性回归模型中，估计中存在随机扰动项，只有对随机扰动的分布等做出假定，才能确定所估计参数的分布性质，也才能够使得做出的估计具有良好的统计性质。这些假定分别是零均值假定、同方差假定、无自相关性假定、随机扰动与解释变量不相关假定、正态性假定和无多重共线性假定。在对模型和变量，以及随机扰动项做出假定的基础上，我们寻找一种理想的估计方法，使得剩余平方和最小，即$Y_i$与$\hat{Y_i}$的差值$e_i$尽可能小，也就是

$$min(\sum{e_i^2}) = min(\sum(Y_i - \hat{Y_i})^2)$$

即

$$min(\sum{e_i^2}) = min(\sum(Y_i - \hat{\beta_1} - \hat{\beta_2}X_{2i} - \hat{\beta_3}X_{3i} - \hat{\beta_4}X_{4i})^2)$$

当然，参数估计与模型检验是一个反复的过程，只有此种估计方法和检验结果表明模型符合古典假定，并满足各项检验，才能采用这种估计结果。否则，必须对模型加以调整，或者考虑采用其他估计方法。

### （三）检验方法

计量经济模型的检验主要包括经济意义检验、统计推断检验、计量经济学检验、模型预测检验等。

经济意义检验主要是检验所估计的模型参数的数值和符号是否符合特定的经济意义。如果所估计的参数与经济理论或实际经验不符，则应思考模型、数据、假定和估计方法有无需要改进的地方，或者经济理论和经验有无不完备的地方。

统计推断检验主要是拟合优度的检验、$F$检验、$t$检验，用以分析模型和各个变量是否显著。在多元回归模型中，先进行$F$检验，对模型整体回归显著性进行检验，若检验结果接受$H_0$则检验解释，表示模型整体不具有显著性，需要重新设置解释变量，反之才能进一步进行$t$检验，对每个解释变量各自的回归系数显著性进行检验。

计量经济学检验，主要是对多重共线性、异方差性、自相关性、设定误差的检验。这些检验其实都是对OLS估计方法中随机扰动项分布假定的检验，其中设定误差检验可以对随机扰动项与解释变量相关关系进行检验，因为各种模型设定误差的最终表现均为随机扰动项与解释变量相关。

如果模型检验存在问题，则需要灵活运用一定的方法加以解决。估计、检验与调整是循环往复、相辅相成、有机结合、螺旋上升的过程。

## 四、数据及处理

### （一）数据搜集

根据国家统计局官网上的信息，虽然没有发布过“中国居民储蓄率”，但居民储蓄率可通过相关数据进行计算（来源：国家统计局《中国统计年鉴》），即

$$居民储蓄率 = {住户部门总储蓄 \over 住户部门可支配收入}$$

其中，$住户部门总储蓄 = 住户部门可支配收入 - 住户部门最终消费$。

具体来说，“住户部门总储蓄”选取《资金流量表》住户部门栏目下的“总储蓄”，“住户部门可支配收入”选取《资金流量表》住户部门栏目下的“调整后可支配总收入”，如果没有则用“可支配总收入”代替，“住户部门最终消费”选取《资金流量表》住户部门栏目下的“居民实际最终消费”。从经济指标的角度，住户指共享同一生活设施，共同使用部分或全部收入和财产，共同消费住房、食品和其他消费品与服务的常住个人或个人群体。所有住户组成住户部门。总储蓄指可支配总收入用于最终消费后的余额。调整后可支配总收入指在各机构部门可支配总收入的基础上，加上该部门应得的实物社会转移，减去该部门应付的实物社会转移，形成调整后可支配总收入。可支配总收入指在初次分配总收入的基础上，通过经常转移的形式对初次分配总收入进行再次分配。居民实际最终消费指常住住户获得的用于消费的货物和服务价值。

居民消费价格指数是反映一定时期内居民所消费商品及服务项目的价格水平变动趋势和变动程度。本课题选取了2000年～2017年的数据（以1978年为100）。（来源：国家统计局）根据公式$通货膨胀率 = {{现期价格指数-基期价格指数}\over 基期价格指数}$就可以根据2000年～2017年的居民消费价格指数计算出2001年～2017年的通货膨胀率。

存款利率是商业银行或类似银行为活期、定期或储蓄类存款支付的利率。本课题选取了2001年～2017年的数据。（来源：国际货币基金组织《国际金融统计》和数据文件）

城镇登记失业率是城镇登记失业人员与城镇单位就业人员(扣除使用的农村劳动力、聘用的离退休人员、港澳台及外方人员)、城镇单位中的不在岗职工、城镇私营业主、个体户主、城镇私营企业和个体就业人员、城镇登记失业人员之和的比。本课题选取了2001年～2017年的数据（据《中国统计年鉴2020》，2018年劳动力数据做过一次调整，可能会带来一定影响）。（来源：国家统计局《中国统计年鉴》）

接下来对我们所选用的时间序列变量进行平稳性检验。我们分别对变量做ADF检验。（见*附录4*）$Y$的ADF检验中，从检验结果来看，在1%、5%、10%显著水平下，单位根检验的临界值分别为-4.331573、-3.23295、-2.7487，t检验统计量=-0.9791699946893014大于相应临界值，不能拒绝原假设，故$Y$存在单位根，是非平稳序列。同理可得$X_2$是平稳序列，$X_3, X_4$是非平稳序列。

```python
print('Y的一阶差分：', adfuller(df['Y'].diff().dropna()))

print('\nX3的一阶差分：', adfuller(df['X3'].diff().dropna()))
print('X3的二阶差分：', adfuller(df['X3'].diff().dropna().diff().dropna()))

print('\nX4的一阶差分：', adfuller(df['X4'].diff().dropna()))
print('X4的二阶差分：', adfuller(df['X4'].diff().dropna().diff().dropna()))
```

下面对同为二阶单整非平稳序列的$X_3,X_4$进行协整检验，这里选用的是单一方程的EG两步法协整检验。

第一步，以$X_3$作为被解释变量，$X_4$作为解释变量，用OLS法对回归方程

$$ X_{3t} = \alpha + \beta_4 X_{4t} + u_t $$

进行估计，得到残差序列$e_t = X_{3t} - \hat{\alpha} - \hat{\beta}_4 X_{4t}$。

第二步，检验$e_t$的平稳性，若$e_t$是平稳的，则$X_3,X_4$是协整的，反之亦然。其中，根据书上的公式和附表可以求得，在5%显著性水平下的临界值为$-3.3377-{5.967\over 17}-{8.98\over 17^2}=-3.7198$。

具体的实现过程及结果见*附录6*。从检验结果来看，t统计量为-2.247656186052834，大于临界值，所以接受原假设，即残差序列是非平稳序列，说明$X_3,X_4$之间不存在协整关系。这意味着模型可能造成伪回归，用设两个变量构造的模型是没有意义的，后续可能要考虑剔除某一变量。

```python
import statsmodels.formula.api as smf

est1 = smf.ols(formula='X3 ~ X4', data=df).fit()
df['e34'] = df['X3'] - est1.predict(df['X4'])
print('残差e的ADF检验：', adfuller(df['e34']))
```

## 五、结果分析

### （一）参数估计

我们先使用OLS法对前面设定的模型进行估计，具体过程见*附录7*。从中我们可以得到

$$ Y = 4.3787 + 0.4625 * X_2 + 1.6237 * X_3 + 6.8152 * X_4 $$

$$ SE = (17.204)\ \ (0.427)\ \ (1.126)\ \ (4.176) $$

$$ t = (0.255)\ \ (1.083)\ \ (1.443)\ \ (1.632) $$

$$ R^2 = 0.451\ \ \ \ \overline{R^2} = 0.324\ \ \ \ F = 3.553\ \ \ \ df = 13 $$

```python
est2 = smf.ols(formula='Y ~ X2 + X3 + X4', data=df).fit()
# 打印回归结果
print(est2.summary())
```

### （二）模型检验

#### 1 拟合优度检验

可决系数表示的是总变差中由模型做出了解释的部分所占的比重，多重可决系数可表示为

$$R^2 = {\sum{(\hat{Y_i}-\overline{Y})^2}\over{\sum{(Y_i-\overline{Y})^2}}} = 1-{{\sum{e_i^2}}\over \sum{y_i^2}}$$

需要注意的是，多重可决系数是模型中解释变量个数的不减函数，在对比不同模型的多重可决系数时会带来缺陷，所以需要修正。修正的可决系数为：

$$\overline{R^2} = 1-{{\sum{e_i^2}/(n-k)}\over \sum{y_i^2}/(n-1)} = 1 - {{n-1}\over{n-k}}{{\sum{e_i^2}}\over \sum{y_i^2}}$$

从上面的结果可以看出，可决系数$R^2 = 0.451$，修正后的可决系数$\overline{R^2} = 0.324$，说明模型对样本的拟合并不好，需要对模型的设定做出调整。

#### 2 F检验

多元回归由于存在多个解释变量，所以需要说明所有解释变量联合起来对被解释变量影响的总体显著性，或整个方程总的联合显著性。这就需要在方差分析的基础上进行F检验。建立统计量

$$F = {{{ESS/(k-1)}\over{RSS/(n-k)}} ～ F(k-1,n-k)}$$

给定显著性水平$\alpha = 0.05$，在F分布中查出自由度为$k-1=3$，$n-k=13$的临界值$F_{0.05}{(3,13)} = 3.41$。而$F = 3.553 > F_{0.05}{(3,13)} = 3.41$，所以应该拒绝原假设$H_0: \beta_2 = \beta_3 = \beta_4 = 0$，说明回归方程显著。

#### 3 t检验

除了检验多个解释变量联合起来对被解释变量的显著性，还需要检验各个解释变量独自对被解释变量的显著性，这就需要分别对每个回归系数逐个地进行t检验。由此我们可以发现，在一元回归中F检验与t检验等价，且$F=t^2$，但在多元回归中F检验与t检验作用不同，故需要分别进行。

若给定显著性水平$\alpha = 0.05$，查t分布表可得自由度为$n-k = 13$时临界值$t_{0.025}(13) = 2.160$；若给定显著性水平$\alpha = 0.10$，查t分布表可得自由度为$n-k = 13$时临界值$t_{0.05}(13) = 1.771$，但是$\beta_2,\beta_3,\beta_4$的t统计量均小于1.771，表明在给定显著性水平0.1的条件下还不能拒绝原假设，说明每个解释变量都不显著，模型亟需调整。

#### 4 经济意义检验

估计后的的多元线性回归模型表示，通货膨胀率每增加1个百分点，居民储蓄率就增加0.4625个百分点，这显然不符合我们的猜想，因为物价上涨储蓄反而增加实在有违常理。银行存款利率每增加1个百分点，居民储蓄率就增加1.6237个百分点，可能有一定道理。失业率每增加1个百分点，居民储蓄率就增加6.8152个百分点，可能比较合理，因为失业率数据总体波动不大，不过一旦有所波动就会对经济社会造成较大影响。

可以发现，模型的拟合优度检验、t检验和经济意义检验都存在问题。下面对模型进行古典假定的检验，即多重共线性、异方差性、自相关性检验。

#### 5. 多重共线性检验

虽然简单相关系数法也能对多重共线性做检验，但高简单相关系数只是多重共线性的充分条件，而非必要条件。所以这里选用方差扩大因子法。对于多元线性回归模型来说，如果分别以每个解释变量为被解释变量，做与其他解释变量的回归，这称为**辅助回归**。用$R_j^2$表示以$X_j$为被解释变量做对其他解释变量辅助线性回归的可决系数，则可以证明解释变量$X_j$参数估计值$\hat\beta_j$的方差可表示为

$$Var(\hat{\beta}_j) = {{\sigma^2}\over{\sum{x_j^2}}}{1\over{1-R_j^2}} = {{\sigma^2}\over{\sum{x_j^2}}}VIF_j$$

其中，$VIF_j$是变量$X_j$的方差扩大因子，即

$$VIF_j = {1\over {1-R_j^2}}$$

由于$R_j^2$度量了$X_j$与其他变量的线性相关程度，这种相关程度越强，说明变量间多重共线性越严重，$VIF_j$也就越大。反之，$R_j^2$越小，说明变量间线性相关性越弱，$VIF_j$也就越接近1。由此可见，我们可以用$VIF_j$作为变量间多重共线性严重程度的度量。

经验表明，$VIF_j \geq 10$时，说明解释变量与其余解释变量之间有严重的多重共线性，且这种多重共线性可能会过度地影响最小二乘估计。VIF的计算过程及结果见*附录8*。检验结果显示，变量间并不存在明显的多重共线性。

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df.iloc[:,2:5]
# 加上常数项
X["c"] = np.ones((len(df)))

for i in range(X.shape[1] - 1):
    print("解释变量X{}的方差扩大因子为{}".format(i+2,round(variance_inflation_factor(X.values,i),2)))
```

#### 6. 异方差性检验

对于异方差性的检验我们选用White检验。White检验不需要关于异方差的任何先验信息，只需要在大样本的情况下，将OLS估计后的残差平方对常数、解释变量、解释变量的平方及其交叉乘积等构成辅助回归，利用辅助回归建立相应的检验统计量来判断异方差性。

White检验的步骤如下：（实现过程及结果见*附录9*）

1. 用OLS估计原模型$Y_i = \beta_1 + \beta_2X_{2i} + \beta_3X_{3i} + \beta_4X_{4i} + u_i$，计算残差$e_t = Y_t - \hat{Y_t}$，并求其平方。

2. 用残差平方$e_t^2$作为异方差$\sigma_t^2$的估计，做辅助函数$e_t^2 = \alpha_1 + \alpha_2X_{2t} + \alpha_3X_{3t} + \alpha_4X_{4t} + \alpha_5X^2_{2t} + \alpha_6X^2_{3t} + \alpha_7X^2_{4t} + \alpha_8X_{2t}X_{3t} + \alpha_9X_{2t}X_{4t} + \alpha_{10}X_{3t}X_{4t} + v_t$，并用OLS对其做估计。

3. 计算统计量$nR^2$。

4. 假设$H_0: \alpha_2 = ... = \alpha_{10} = 0$，则可证明$nR^2$渐进服从自由度为$p$的$\chi^2$分布，$p=9$。给定显著性水平$\alpha = 0.05$时，$nR^2 > \chi_{0.05}^2(9)$，所以拒绝原假设，即表明模型存在异方差。

```python
df['e2'] = (df['Y'] - est2.predict(df.iloc[:,2:5]))**2
est3 = smf.ols(formula='e2 ~ X2 + X3 + X4 + np.sqrt(X2) + np.sqrt(X3) + np.sqrt(X4) + X2*X3 + X2*X4 + X3*X4', data=df).fit()
print(est3.summary())
```

### （三）模型调整

由于模型存在异方差，所以我们需要对模型设定及参数估计做出一些调整。最常采用的方法是加权最小二乘法。

加权最小二乘的想法是，对于不同的样本给予不同的权重，即对较小的$e_i^2$给予较大的权数，对较大的$e_i^2$给予较小的权数。通常可将权数取为$w_i = {1\over\sigma_i^2}\ \ (i = 1,2,...,n$)，将权数与残差平方和相乘后再求和，得

$$\sum{w_ie_i^2} = \sum{w_i(Y_i-\hat{\beta_1}-\hat{\beta_2}X_{2i}-\hat{\beta_3}X_{3i}-\hat{\beta_4}X_{4i})^2}$$

称为加权的残差平方和。根据最小二乘原理，即要使得

$$min\sum{w_ie_i^2} = min\sum{w_i(Y_i-\hat{\beta_2}X_{2i}-\hat{\beta_3}X_{3i}-\hat{\beta_4}X_{4i})^2}$$

具体的实现过程及结果见*附录10*。观察后可以发现，WLS估计的参数均可以在显著性水平为5%的条件下显著，相较于原来有了非常大的提升。但是在检验经济意义的时候，$\beta_2>0$即通货膨胀率每增加1个百分点，居民储蓄率就增加0.5981个百分点，还是不符合我们的预期。通过学习一些经济学知识后，我们做出猜想，即通货膨胀导致人民币贬值，债务人偿还原来欠下的债务的压力减小，留下更多的钱可用于储蓄。因此可以看出，通货膨胀和储蓄率之间的关系是复杂而微妙的。

```python
import statsmodels.api as sm

est4 = sm.WLS(df['Y'],sm.add_constant(df.iloc[:,2:5]),weights = 1/df['e2']).fit()

# 打印回归结果
print(est4.summary())
```

另一方面，从数据分析和处理中可以发现，$X_3,X_4$是不协整的，所以我们需要剔除其中一个变量；$X_4$的变异系数较小，可能不显著。因此比较合理的方式是把变量$X_4$剔除，然后对剩下的变量继续使用WLS（估计结果见*附录11*），可以发现变量的显著性得到了进一步提升。模型最后的估计结果为

$$ Y = 33.0814 + 0.6627 * X_2 + 1.2115 * X_3 $$

$$ SE = (0.684)\ \ (0.118)\ \ (0.303) $$

$$ t = (48.390)\ \ (5.608)\ \ (3.996) $$

$$ R^2 = 0.908\ \ \ \ \overline{R^2} = 0.895\ \ \ \ F = 69.07\ \ \ \ df = 14 $$

调整后的模型可决系数$R^2 = 0.908$，修正后的可决系数$\overline{R^2} = 0.895$，说明模型对样本的拟合得比较好。

给定显著性水平$\alpha = 0.05$下，$F = 69.07 > F_{0.05}{(2,14)}$，所以拒绝原假设，回归方程显著。

若给定显著性水平$\alpha = 0.05$，查t分布表可得自由度为$n-k = 14$时临界值$t_{0.025}(14) = 2.145$，所以$\beta_2, \beta_3$的t统计量均大于2.145，拒绝原假设，$X_2, X_3$均显著。

估计后的的多元线性回归模型表示，通货膨胀率每增加1个百分点，居民储蓄率就增加0.6627个百分点，银行存款利率每增加1个百分点，居民储蓄率就增加1.2115个百分点。

```python
est5 = smf.ols(formula='Y ~ X2 + X3', data=df).fit()
df['e2_'] = (df['Y'] - est5.predict(df.iloc[:,2:4]))**2
est6 = sm.WLS(df['Y'],sm.add_constant(df.iloc[:,2:4]),weights = 1/df['e2_']).fit()

# 打印回归结果
print(est6.summary())
```
